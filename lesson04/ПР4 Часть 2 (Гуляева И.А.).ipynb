## Задание 9

Создать одномерный массив _**а**_, состоящий из последовательности целых чисел от 0 до 12. Поменять форму этого массива, чтобы получилась матрица _**A**_ (двумерный массив **Numpy**), состоящая из 4 строк и 3 столбцов. Получить матрицу _**At**_ путём транспонирования матрицы _**A**_. Получить матрицу _**B**_, умножив матрицу _**A**_ на матрицу _**At**_ с помощью матричного умножения. Какой размер имеет матрица _**B**_?

Получится ли вычислить обратную матрицу для матрицы _**B**_ и почему?

```python
a = np.arange(12, dtype = 'int32')
print('Массив a: {}'.format(a))
print('Форма массива a: {}'.format(a.shape))
print('Размерность массива a: {}'.format(a.ndim))
print('Количество элементов массива a: {}'.format(a.size))
```

```python
A = a.reshape(4, 3)
print('Матрица A:\n', A)
print('Форма матрицы A: {}'.format(A.shape))
print('Размерность матрицы A: {}'.format(A.ndim))
print('Количество элементов матрицы A: {}'.format(A.size))
```

```python
At = A.T
print('Транспонированная матрица A:\n', At)
print('Форма транспонированной матрицы A: {}'.format(At.shape))
print('Размерность транспонированной матрицы A: {}'.format(At.ndim))
print('Количество элементов транспонированной матрицы A: {}'.format(At.size))
```

```python
B = A.dot(At)
print('Матрица B:\n', B)
print('Форма матрицы B: {}'.format(B.shape))
print('Размер матрицы B: {}'.format(len(B)))
print('Ранг матрицы B: {}'.format(np.linalg.matrix_rank(B)))
print('Определитель матрицы B: {}'.format(np.linalg.det(B)))
```

### Ответ:

Вычислить обратную матрицу для квадратной мартицы _**B**_ не получится, в виду нулевого значения её определителя. А если определитель матрицы равен 0, то определитль обратной матрицы равен 1/0 (что недопустимо), именно поэтому матрицы с нулевым определителем обращать нельзя.

Определитель матрицы отличен от 0 тогда и только тогда, когда все строки (вектора) матрицы являются линейно независимыми. Ленейная независимость векторов матрицы определяется равенством её размера и ранга. И так как в нашем случае ранг (2) не равен размеру (4) марицы, мы получили нулевой определитель.

## Задание 12

Поменяйте форму массива _**c**_ из Задания 11 так, чтобы получилась квадратная матрица _**C**_. Получите матрицу _**D**_, поэлементно прибавив матрицу _**B**_ из Задания 9 к матрице _**C**_, умноженной на 10. Вычислите определитель, ранг и обратную матрицу **D_inv** для _**D**_.

```python
C = c.reshape(4, 4)
print('Матрица C:\n', C)
print('Тип матрицы C: {}'.format(C.dtype))
print('Форма матрицы C: {}'.format(C.shape))
print('Размерность матрицы C: {}'.format(C.ndim))
print('Количество элементов матрицы C: {}'.format(C.size))
```

```python
D = B + C * 10
print('Матрица D:\n', D)
print('Тип матрицы D: {}'.format(D.dtype))
print('Форма матрицы D: {}'.format(D.shape))
print('Размер матрицы D: {}'.format(len(D)))
print('Ранг матрицы D: {}'.format(np.linalg.matrix_rank(D)))
print('Определитель матрицы D: {}'.format(np.linalg.det(D)))
```

```python
D_inv = np.linalg.inv(D)
print('Обратная матрица D_inv:\n', D_inv)
print('Тип обратной матрицы D_inv: {}'.format(D_inv.dtype))
print('Форма обратной матрицы D_inv: {}'.format(D_inv.shape))
print('Размер матрицы D_inv: {}'.format(len(D_inv)))
print('Ранг матрицы D_inv: {}'.format(np.linalg.matrix_rank(D_inv)))
print('Определитель матрицы D_inv: {}'.format(np.linalg.det(D_inv)))
```

Проверка обратной матрицы производится путём умножения  её на исходную матрицу, что в результате должно дать единичную матрицу. Убедимся в этом:

```python
print('Проверочная единичная матрица:\n', D.dot(D_inv))
```

С помощью функции **numpy.where**, используя матрицу _**D_inv_new**_ в качестве маски, а матрицы _**B**_ (из Задания 9) и _**C**_ (из Задания 12) - в качестве источников данных, получите матрицу _**E**_ размером 4x4. Элементы матрицы _**E**_, для которых соответствующий элемент матрицы _**D_inv_new**_ равен 1, должны быть равны соответствующему элементу матрицы _**B**_, а элементы матрицы _**E**_, для которых соответствующий элемент матрицы _**D_inv_new**_ равен 0, должны быть равны соответствующему элементу матрицы _**C**_.

```python
D_inv_new = np.where(D_inv < 0, 0, 1)
print('Матрица D_inv_new:\n', D_inv_new)
print('Тип матрицы D_inv_new: {}'.format(D_inv_new.dtype))
print('Форма матрицы D_inv_new: {}'.format(D_inv_new.shape))
```

```python
print('Матрица B:\n', B)
print('Тип матрицы B: {}'.format(B.dtype))
print('Форма матрицы B: {}'.format(B.shape))
```

```python
print('Матрица C:\n', C)
print('Тип матрицы C: {}'.format(C.dtype))
print('Форма матрицы C: {}'.format(C.shape))
```
