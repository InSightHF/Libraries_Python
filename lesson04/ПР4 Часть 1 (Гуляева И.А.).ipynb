
## Задание 2

С помощью функции **linspace** из библиотеки **Numpy** создайте массив _**t**_ из 51 числа от 0 до 10 включительно. Создайте массив **Numpy** под названием _**f**_, содержащий косинусы элементов массива _**t**_. Постройте линейную диаграмму, используя массив _**t**_ для координат по горизонтали, а массив _**f**_ - для координат по вертикали. Линия графика должна быть зелёного цвета. Выведите название диаграммы - **'График f(t)'**. Также добавьте названия для горизонтальной оси - 'Значения t' и для вертикальной - 'Значения f'. Ограничьте график по оси _**x**_ значениями 0.5 и 9.5, а по оси _**y**_ - значениями -2.5 и 2.5.


```python
t = np.linspace(0, 10, 51)
print(t)
```

```python
f = np.cos(t)
print(f)
```

```python
plt.plot(t, f, color = 'green')
plt.title('График f(t)', fontsize = 14, fontweight = 'bold')
plt.xlabel('Значения t', fontweight = 'bold')
plt.ylabel('Значения f', fontweight = 'bold')
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show()
```


## Задание 3*

С помощью функции **linspace** библиотеки **Numpy** создайте массив x из 51 числа от -3 до 3 включительно. Создайте массивы _**y1, y2, y3, y4**_ по следующим формулам: **y1 = x^2; y2 = 2 * x + 0.5; y3 = -3 * x - 1.5; y4 = sin(x)**. Используя функцию **subplots** модуля **matplotlib.pyplot**, создайте объект **matplotlib.figure.Figure** с названием **fig** и массив объектов **Axes** под названием **ax**, причём так, чтобы у вас было 4 отдельных графика в сетке, состоящей из двух строк и двух столбцов. В каждом графике массив _**x**_ используется для координат по горизонтали. В левом верхнем графике для координат по вертикали используйте _**y1**_, в правом верхнем - _**y2**_, в левом нижнем - _**y3**_, в правом нижнем - _**y4**_. Дайте название графикам: **'График y1', 'График y2'** и т.д. Для графика в левом верхнем углу установите границы по оси _**x**_ от -5 до 5. Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали. Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3.


```python
x = np.linspace(-3, 3, 51)
print(x)
```

```python
y1 = x**2
print(y1)
```

```python
y2 = 2 * x + 0.5
print(y2)
```

```python
y3 = -3 * x - 1.5
print(y3)
```

```python
y4 = np.sin(x)
print(y4)
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2)
fig.set_size_inches(8, 6)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x, y1)
ax1.set_xlim([-5, 5])
ax1.set_title('График y1', fontweight = 'bold')
ax2.plot(x, y2)
ax2.set_title('График y2', fontweight = 'bold')
ax3.plot(x, y3)
ax3.set_title('График y3', fontweight = 'bold')
ax4.plot(x, y4)
ax4.set_title('График y4', fontweight = 'bold')
plt.subplots_adjust(wspace = 0.3, hspace = 0.3)
plt.show()
```

На следующем графике постройте две гистограммы по значениям признака **V1** - одну для мошеннических транзакций (Class равен 1) и другую - для обычных (Class равен 0). Создайте легенду с двумя значениями: **“Class 0”** и **“Class 1”**. Гистограмма обычных транзакций должна быть серого цвета, а мошеннических - красного. Горизонтальной оси дайте название **“V1”**.

```python
import pandas as pd
plt.style.use('fivethirtyeight')
```

```python
DATASET_PATH = './creditcard.csv'
creditcard = pd.read_csv(DATASET_PATH)
print(creditcard.shape)
```

```python
class_variable = creditcard['Class'].value_counts()
print(class_variable)
```

```python
class_variable.plot(kind = 'barh', color='violet', ec = 'black', alpha=0.7)
plt.title('Столбчатая диаграмма class_variable', fontsize = 14, fontweight = 'bold', color = 'blue', family = 'serif')
plt.xlabel('Value_counts')
plt.ylabel('Class')
plt.show()
```

```python
class_variable.plot(kind = 'barh', logx=True, color='yellow', ec = 'black', alpha=0.7)
plt.title('Столбчатая диаграмма class_variable', fontsize = 14, fontweight = 'bold', color = 'blue', family = 'serif')
plt.xlabel('Value_counts (log)')
plt.ylabel('Class')
plt.show()
```

```python
class_0 = creditcard.loc[creditcard['Class'] == 0, ['V1']]
print(class_0.describe())
```

```python
class_1 = creditcard.loc[creditcard['Class'] == 1, ['V1']]
print(class_1.describe())
```

```python
plt.hist(class_0['V1'], bins=20, ec = 'black', density=True, alpha=0.5, label='Class_0', color='grey')
plt.hist(class_1['V1'], bins=20, ec = 'white', density=True, alpha=0.5, label='Class_1', color='red')
plt.title('Гистограммы признака V1', fontsize = 16, fontweight = 'bold', color = 'orange', family = 'serif', loc = 'left')
plt.xlabel('V1')
plt.ylabel('Density')
plt.axis([-33, 3, 0.00, 0.23])
plt.legend(frameon = False)
plt.show()
```
